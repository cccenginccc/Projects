```python
A_square = 4*(1**2)
```

```python
A_circle = (m.pi)*(1**2) 
```

E)
Here, I add a random noise as described, however since the colour scale is 0-256 in my format, I encountered that values higher than 255 are displayed as white in the image, therefore, I decided to apply modulo function to the formatted pixels so that new values remains in the defined range

```python
noisy_image = np.empty((512,512,3))
for i in range(0,512):
    for j in range(0,512):
        noisy_image[i,j,0] = np.around(image[i,j,0] + 0.1 * np.random.random_sample(size = 1)*image[:,:,0].max())%256
        noisy_image[i,j,1] = np.around(image[i,j,1] + 0.1 * np.random.random_sample(size = 1)*image[:,:,1].max())%256
        noisy_image[i,j,2] = np.around(image[i,j,2] + 0.1 * np.random.random_sample(size = 1)*image[:,:,2].max())%256
```

```python
plt.imshow(noisy_image)
```
